```python
from probabilit.modeling import Distribution
```

```python
thickness = Distribution("norm",loc=50, scale=10)
```

```python
area = Distribution("norm",loc=100, scale=20)
```

```python
porosity = Distribution("lognorm", s=0.1, scale=0.1)
```

```python
addntg = Distribution("norm", loc=0.6, scale=0.1)
```

```python
gpv = thickness * area * porosity * addntg
```

```python
gpv
```

```python
samples = gpv.sample(200, random_state=42)
samples
```

```python
import matplotlib.pyplot as plt

plt.hist(samples, bins=30, edgecolor='black')
plt.xlabel('GPV')
plt.ylabel('Frequency')
plt.title('Distribution of GPV')
plt.show()
```
